```python
import pandas as pd
import matplotlib as mp
```

```python
votes = pd.read_csv("votes.csv", index_col=0, usecols=["id", "vote_id", "voting_choice_id", "alternative_ordinal"])
```

```python
candidates = {
    2: "Rowey",
    3: "Picoo",
    4: "biass",
    5: "GFY_DePonny",
    6: "FtXCommando"
}
```

```python
def runoff(all_votes, candidate, ordinal=0):
    """
    Perform AV-runoff for the given candidate
    """
    votes_to_discard = all_votes.query("voting_choice_id == %i and alternative_ordinal == %i" % (candidate, ordinal))
    
    new_votes = all_votes.merge(votes_to_discard, on="vote_id").query("alternative_ordinal_x == %s" % str(ordinal+1))
    return (
        new_votes
            .rename(index=int,
                    columns={"alternative_ordinal_y": "alternative_ordinal", "voting_choice_id_x": "voting_choice_id"})
            .drop(labels=["voting_choice_id_y", "alternative_ordinal_x"], axis=1)
    ).groupby(["voting_choice_id"])["vote_id"].count()

def candidate_plot(input, title):
    plt = (input.plot.bar(
        title=title))
    plt.set_xlabel("Candidate")
    plt.set_xticklabels(candidates.values())
    return plt
    
primary_votes = votes.query("alternative_ordinal == 0")
primary_picks = primary_votes.groupby(["voting_choice_id"])["vote_id"].count()
candidate_plot(primary_picks, "PC Vote First Round")
```

```python
runoff_1 = runoff(votes, 3)
del candidates[3]
candidate_plot(runoff_1, "First runoff")
```

```python
round_2 = primary_picks.drop(3).add(runoff_1)
candidate_plot(round_2, "PC Vote Second Round")
```

```python
#del candidates[5]
runoff_2 = runoff(votes, 5).drop(3)
candidate_plot(runoff_2, "Second runoff")
```

```python
round_3 = round_2.drop(5).add(runoff_2)
candidate_plot(round_3, "PC Vote Third Round")
```

```python
#del candidates[2]
runoff_3 = runoff(votes, 2).drop(3).drop(5)
candidate_plot(runoff_3, "Third runoff")
```

```python
round_4 = round_3.add(runoff_3).drop(2)
candidate_plot(round_4, "PC Vote Final Round")
```
